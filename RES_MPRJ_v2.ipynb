```python
# PEGA LINKS PÁGINA

arrLinks = navegador.find_elements_by_xpath("//div[@class='content-card']/a")

itensPaginaLink = pd.DataFrame(data=arrLinks, columns=["ObjetoLink"])
#itensPaginaLink.head()
itensPaginaLink["Link1"] = itensPaginaLink.applymap(lambda x: x.get_attribute("href"))
display(itensPaginaLink)
```

```python
# PEGA TITULOS PÁGINA

arrTitulos = navegador.find_elements_by_xpath("//h3")
print(len(arrTitulos))
itensPaginaTit = pd.DataFrame(data=arrTitulos, columns=["ObjetoTitulo"])
itensPaginaTit.head()
itensPaginaTit["Titulo1"] = itensPaginaTit.applymap(lambda x: x.text)
itensPaginaTit.head()
display(itensPaginaTit)
```

```python
# PEGA TEXTOS PÁGINA


arrTextos = navegador.find_elements_by_xpath("//div[@class='content-card']/a/div/div/div")

print(len(arrTextos))
itensPaginaTexto = pd.DataFrame(data=arrTextos, columns=["ObjetoTexto"])

itensPaginaTexto["Texto1"] = itensPaginaTexto.applymap(lambda x: x.text)
itensPaginaTexto.head(10)
#display(itensPaginaTexto.shape)
```

```python
arrITEM = pd.DataFrame()
arrITEM["Link"] = itensPaginaLink["Link1"]
arrITEM["Titulo"] = itensPaginaTit["Titulo1"]
arrITEM["Texto"] = itensPaginaTexto["Texto1"]
```
